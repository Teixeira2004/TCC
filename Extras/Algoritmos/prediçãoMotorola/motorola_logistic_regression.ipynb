```python
# Importando as bibliotecas necessárias
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
```

```python
# Carregando o arquivo CSV da Asus
file_path = 'gdrive/My Drive/dados_smartphones_v4_Motorola_tratado.csv'
df_motorola = pd.read_csv(file_path, sep=";")
```

```python
print(df_motorola.columns)
```

```python
# Dividindo a variavel dependnete (Selling Price Dolar) -y- e as variáveis independentes (Memory e Storage) -x-
y = df_motorola['selling_price_dolar']
x = df_motorola.drop('selling_price_dolar',axis=1)
```

```python
# Codificando a variável dependente Selling Price Dolar em categorias numéricas
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Verificando o número de amostras disponíveis após o processamento
print("\nTotal de amostras disponíveis para treino e teste:", len(x))
```

```python
# Dividindo o conjunto de dados em treino e teste
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size=0.3)
```

```python
# Treinando o modelo de Regressão Logística
model_lr = LogisticRegression(random_state=1986)
model_lr.fit(x_treino, y_treino)
```

```python
# Previsões e avaliação do modelo
y_pred_lr = model_lr.predict(x_teste)
accuracy_lr = accuracy_score(y_teste, y_pred_lr)
classification_report_lr = classification_report(y_teste, y_pred_lr, target_names=label_encoder.classes_.astype(str))
```

```python
# Exibindo os resultados
print("\nAcurácia do modelo LogisticRegression:", accuracy_lr)
print("\nRelatório de Classificação do modelo LogisticRegression:\n", classification_report_lr)
```
